```python
import pandas as pd
from collections import Counter

import matplotlib as mpl 
import matplotlib.pyplot as plt 

import numpy as np
import seaborn

datum_twor = np.array(pd.read_csv("twor_uncertain_mean.csv",header=None).values)
datum_two  = np.array(pd.read_csv("two_uncertain_mean.csv",header=None).values)
datum_mixture  = np.array(pd.read_csv("Gaussian_mixture.csv",header=None).values)

samplesize = ["1","5","10","15","20","25","30","35","40","45","50"]
u_0 = ["0","0.5","1","1.5","2","2.5","3","3.5","4","4.5","5","5.5","6"]
u_twor = ["0","0.5","1","1.5","2","2.5","3","3.5","4","4.5","5","5.5","6","100"]
```

```python
twor_mean = datum_twor[0:14,::2]
twor_std = datum_twor[0:14,1::2]
twor_fit = datum_twor[14:28:,::2]
twor_predict = datum_twor[28:,::2]

two_mean = datum_two[0:13,::2]
two_std = datum_two[0:13,1::2]
two_fit = datum_two[13:26:,::2]
two_predict = datum_two[26:,::2]

m_2 = datum_mixture[1::9,0]
m_3 = datum_mixture[2::9,0]
m_2_std = datum_mixture[1::9,1]
m_3_std = datum_mixture[2::9,1]
m_2_fit = datum_mixture[4::9,0]
m_3_fit = datum_mixture[5::9,0]
m_2_pre = datum_mixture[7::9,0]
m_3_pre = datum_mixture[8::9,0]
```

# 混合高斯

```python
m_mean = np.vstack([two_mean[0,:],m_2,m_3])
m_std = np.vstack([two_std[0,:],m_2_std,m_3_std])
m_fit = np.vstack([two_fit[0,:],m_2_fit,m_3_fit])
m_predict = np.vstack([two_predict[0,:],m_2_pre,m_3_pre])
```

```python
plt.figure(figsize=(15,7)) 
plt.subplot(2,1,1)
seaborn.heatmap(m_mean,annot=True,xticklabels=samplesize,yticklabels=["1","2","3"],fmt=".4f",cmap="Reds",annot_kws={"size":10,"weight":"bold"})

plt.subplot(2,1,2)
seaborn.heatmap(m_std/m_mean,annot=True,xticklabels=samplesize,yticklabels=["1","2","3"],fmt=".4f",cmap="Blues",annot_kws={"size":10,"weight":"bold"})
```

```python
plt.figure(figsize=(15,7)) 
plt.subplot(2,1,1)
seaborn.heatmap(m_fit,annot=True,xticklabels=samplesize,yticklabels=["1","2","3"],fmt=".4f",cmap="Greens",annot_kws={"size":10,"weight":"bold"})

plt.subplot(2,1,2)
seaborn.heatmap(m_predict,annot=True,xticklabels=samplesize,yticklabels=["1","2","3"],fmt=".4f",cmap="Greens",annot_kws={"size":10,"weight":"bold"})
```

# 高斯

```python
plt.figure(figsize=(15,10)) 
plt.subplot(2,1,1)
seaborn.heatmap(two_mean.T,annot=True,xticklabels=u_0,yticklabels=samplesize,fmt=".4f",cmap="Reds",annot_kws={"size":10,"weight":"bold"})

plt.subplot(2,1,2)
seaborn.heatmap((two_std/two_mean).T,annot=True,xticklabels=u_0,yticklabels=samplesize,fmt=".4f",cmap="Blues",annot_kws={"size":10,"weight":"bold"})
```

```python
plt.figure(figsize=(15,10)) 
plt.subplot(2,1,1)
seaborn.heatmap(two_fit.T,annot=True,xticklabels=u_0,yticklabels=samplesize,fmt=".2f",cmap="Greens",annot_kws={"size":9,"weight":"bold"})

plt.subplot(2,1,2)
seaborn.heatmap(two_predict.T,annot=True,xticklabels=u_0,yticklabels=samplesize,fmt=".2f",cmap="Greens",annot_kws={"size":9,"weight":"bold"})
```

# 高斯回归

```python
plt.figure(figsize=(15,10)) 
plt.subplot(2,1,1)
seaborn.heatmap(twor_mean.T,annot=True,xticklabels=u_twor,yticklabels=samplesize,fmt=".4f",cmap="Reds",annot_kws={"size":10,"weight":"bold"})

plt.subplot(2,1,2)
seaborn.heatmap((twor_std/twor_mean).T,annot=True,xticklabels=u_twor,yticklabels=samplesize,fmt=".4f",cmap="Blues",annot_kws={"size":9,"weight":"bold"})
```

```python
plt.figure(figsize=(15,10)) 
plt.subplot(2,1,1)
seaborn.heatmap((twor_fit).T,annot=True,xticklabels=u_twor,yticklabels=samplesize,fmt=".2f",cmap="Greens",annot_kws={"size":9,"weight":"bold"})

plt.subplot(2,1,2)
seaborn.heatmap((twor_predict).T,annot=True,xticklabels=u_twor,yticklabels=samplesize,fmt=".2f",cmap="Greens",annot_kws={"size":9,"weight":"bold"})
```
