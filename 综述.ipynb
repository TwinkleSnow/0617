```python
import math
import datetime
from time import time
import pandas as pd
from collections import Counter

import matplotlib as mpl 
import matplotlib.pyplot as plt 

import numpy as np
import seaborn

datum_twor = np.array(pd.read_csv("twor_uncertain_mean.csv",header=None).values)
datum_two  = np.array(pd.read_csv("two_uncertain_mean.csv",header=None).values)
datum_mixture  = np.array(pd.read_csv("Gaussian_mixture.csv",header=None).values)
```

```python
twor_mean = datum_twor[0:14,::2]
twor_std = datum_twor[0:14,1::2]
twor_fit = datum_twor[14:28:,::2]
twor_predict = datum_twor[28:,::2]

two_mean = datum_two[0:13,::2]
two_std = datum_two[0:13,1::2]
two_fit = datum_two[13:26:,::2]
two_predict = datum_two[26:,::2]

m_2 = datum_mixture[1::9,0]
m_3 = datum_mixture[2::9,0]
m_2_std = datum_mixture[1::9,1]
m_3_std = datum_mixture[2::9,1]
m_2_fit = datum_mixture[4::9,0]
m_3_fit = datum_mixture[5::9,0]
m_2_pre = datum_mixture[7::9,0]
m_3_pre = datum_mixture[8::9,0]

samplesize = ["1","5","10","15","20","25","30","35","40","45","50"]
```

```python
plt.figure(figsize=(10,5)) 
plt.plot(samplesize,twor_mean[-1,:],color="dodgerblue",marker="X",markersize=10)
plt.plot(samplesize,twor_mean[0,:],color="crimson",marker="^",markersize=10)
plt.plot(samplesize,two_mean[1,:],color="green",marker="X",markersize=10)
plt.plot(samplesize,two_mean[0,:],color="orange",marker="^",markersize=10)
plt.plot(samplesize,m_2,marker="^",markersize=10)
plt.plot(samplesize,m_3,marker="+",markersize=10)
plt.grid()
```

```python
plt.figure(figsize=(10,5)) 
plt.plot(samplesize,twor_std[-1,:]/twor_mean[-1,:],color="dodgerblue",marker="X",markersize=10)
plt.plot(samplesize,twor_std[0,:]/twor_mean[0,:],color="crimson",marker="^",markersize=10)
plt.plot(samplesize,two_std[1,:]/two_mean[1,:],color="green",marker="X",markersize=10)
plt.plot(samplesize,two_std[0,:]/two_mean[0,:],color="orange",marker="^",markersize=10)
plt.plot(samplesize,m_2_std/m_2,marker="^",markersize=10)
plt.plot(samplesize,m_3_std/m_3,marker="+",markersize=10)
plt.axhspan(ymin=0.0,ymax=0.0005,alpha=0.25,color="blueviolet")
plt.grid()
```

```python
plt.scatter("twor",np.std(twor_mean[-1,:])/np.mean(twor_mean[-1,:]))
plt.scatter("oner",np.std(twor_mean[0,:])/np.mean(twor_mean[0,:]))
plt.scatter("two",np.std(two_mean[1,:])/np.mean(two_mean[1,:]))
plt.scatter("one",np.std(two_mean[0,:])/np.mean(two_mean[0,:]))

plt.scatter("m2",np.std(m_2)/np.mean(m_2))
plt.scatter("m3",np.std(m_3)/np.mean(m_3))
plt.grid()
#plt.ylim(0,0.00075)
```

```python
plt.figure(figsize=(10,5)) 
plt.plot(samplesize,twor_fit[-1,:],color="dodgerblue",marker="X",markersize=10)
plt.plot(samplesize,twor_fit[0,:],color="crimson",marker="^",markersize=10)
plt.plot(samplesize,two_fit[1,:],color="green",marker="X",markersize=10)
plt.plot(samplesize,two_fit[0,:],color="orange",marker="^",markersize=10)
plt.plot(samplesize,m_2_fit,marker="^",markersize=10)
plt.plot(samplesize,m_3_fit,marker="+",markersize=10)
plt.grid()
```

```python
plt.figure(figsize=(10,5)) 
plt.plot(samplesize,twor_predict[-1,:],color="dodgerblue",marker="X",markersize=10)
plt.plot(samplesize,twor_predict[0,:],color="crimson",marker="^",markersize=10)
plt.plot(samplesize,two_predict[1,:],color="green",marker="X",markersize=10)
plt.plot(samplesize,two_predict[0,:],color="orange",marker="^",markersize=10)
plt.plot(samplesize,m_2_pre,marker="^",markersize=10)
plt.plot(samplesize,m_3_pre,marker="+",markersize=10)
plt.grid()
```
